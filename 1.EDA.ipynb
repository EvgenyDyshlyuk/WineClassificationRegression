We can check it's predictive capability later. For now let's just change it to index

```python
# Lets restore train index order
df_train = df_train.sort_values(by=['Unnamed: 0'])
df_train = df_train.set_index('Unnamed: 0')
df_test = df_test.set_index('Unnamed: 0')
display(df_train.tail(), df_test.tail())
```

```python
# The train-test split is well balanced for target variable:
print(df_train.quality.value_counts(),
      df_test.quality.value_counts())
```

```python
# Lets see the train and test distributions:
nrows = 2
ncols = 6
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*5,nrows*5))

for i, column in enumerate(df_train.columns):
    sns.distplot(df_train[column],ax=axes[i//ncols,i%ncols], bins = 20, color='b')
    sns.distplot(df_test[column],ax=axes[i//ncols,i%ncols], bins = 20, color='r')
```

- We can see that train and test splits are good, because the distributions are similar. Now we can leave test set aside and continue with the train set

```python
# Let us build a pairplot for all features
sns.pairplot(df_train, hue="quality")
```

- From pairplot we can see that classes are not easily separable. "alcohol" and "volatile acidity" features possibly have the best predictive capability
